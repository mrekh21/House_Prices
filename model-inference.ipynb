```python
import mlflow
```

```python
X_test = pd.read_csv("/kaggle/input/house-prices-advanced-regression-techniques/test.csv")
```

#### დავა-load-ოთ საბოლოო მოდელი mlflow-დან, ტესტ სეტის prediction-ები csv ფაილად შევინახოთ და ავტვირთოთ competition-ზე.

```python
mlflow.set_tracking_uri("https://dagshub.com/mrekh21/House_Prices.mlflow")

logged_model = "models:/HousePricePipeline/5"

loaded_model = mlflow.pyfunc.load_model(logged_model)
```

```python
predictions = loaded_model.predict(X_test)

submission = pd.DataFrame({
    'Id': X_test['Id'],  
    'SalePrice': predictions  
})

submission.to_csv('submission.csv', index=False)
```
